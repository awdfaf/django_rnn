```python
# 데이터프레임 패키지
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import seaborn as sns
import numpy as np

# seaborn 내장 데이터 읽어오기
# 데이터 셋 : anscombe
# 변수이름 : ans
ans = sns.load_dataset("anscombe")
ans.info()
```

```python
# dataset 컬럼의 값을 기준으로 그룹화
# 변수명 : ans_group
ans_group = ans.groupby(["dataset"])
ans_group.head()
```

```python
# 기초통계확인
ans_group.describe()
```

```python
# 시각화 라이브러리
import matplotlib.pylab as plt

# ans 데이터 원본을 이용해 
# dataset 컬럼의 데이터 중 I 값만 추출
data1 = ans[ans["dataset"] == "I"]
data1
```

```python
# dataset = I에 대한 그래프 그리기
plt.plot(data1["x"],data1["y"], "o")
```

```python
# dataset 다른 데이터에 대해서도 필터링 후 그래프 그리기
data2 = ans[ans["dataset"] == "II"]
data3 = ans[ans["dataset"] == "III"]
data4 = ans[ans["dataset"] == "IV"]
```

```python
# 그래프 그리기 위한 객체 받아오기
fig = plt.figure()

# 4개 각각의 그래프를 그리기 위한 공간만들기
# subplot
# 첫번째 값 : 행
# 두번째 값 : 열
# 세번째 값 : 행렬 중 어느 위치에 들어갈지 순서 지정
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

# subplot에 제목넣기
ax1.set_title("data1")
ax2.set_title("data2")
ax3.set_title("data3")
ax4.set_title("data4")



# 데이터 넣어서 그래프 그리기 : 점으로 표시
ax1.plot(data1["x"],data1["y"], "o",c="r")
ax2.plot(data2["x"],data2["y"], "o",c="g")
ax3.plot(data3["x"],data3["y"], "o",c="b")
ax4.plot(data4["x"],data4["y"], "o",c="y")

# 전체 그래프 제목
fig.suptitle("Anscombe Data")

# 그래프 간 간격 조절
# tight_layout()
fig.tight_layout()

# 우상향 데이터 /우하향 데이터 
```

```python
# 데이터프레임 패키지
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import seaborn as sns
import numpy as np
# 시각화 라이브러리
import matplotlib.pylab as plt

# seaborn 내장 데이터 읽어오기
# 데이터 셋 : anscombe
# 변수이름 : ans
ans = sns.load_dataset("anscombe")

# ans 데이터 원본을 이용해 
# dataset 컬럼의 데이터 중 I 값만 추출
data1 = ans[ans["dataset"] == "I"]
# dataset 다른 데이터에 대해서도 필터링 후 그래프 그리기
data2 = ans[ans["dataset"] == "II"]
data3 = ans[ans["dataset"] == "III"]
data4 = ans[ans["dataset"] == "IV"]

# 그래프 그리기 위한 객체 받아오기
fig = plt.figure()

# 4개 각각의 그래프를 그리기 위한 공간만들기
# subplot
# 첫번째 값 : 행
# 두번째 값 : 열
# 세번째 값 : 행렬 중 어느 위치에 들어갈지 순서 지정
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

# subplot에 제목넣기
ax1.set_title("data1")
ax2.set_title("data2")
ax3.set_title("data3")
ax4.set_title("data4")



# 데이터 넣어서 그래프 그리기 : 점으로 표시
ax1.plot(data1["x"],data1["y"], "o",c="r")
ax2.plot(data2["x"],data2["y"], "o",c="g")
ax3.plot(data3["x"],data3["y"], "o",c="b")
ax4.plot(data4["x"],data4["y"], "o",c="y")

# 전체 그래프 제목
fig.suptitle("Anscombe Data")

# 그래프 간 간격 조절
# tight_layout()
fig.tight_layout()

# 완료된 시각화 그래프 이미지 저장
# 저장위치 : images
# 파일명 : anscombe_data.png
```
